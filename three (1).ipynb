```python
import pandas as pd
import numpy as np

# Original Data
data = {
    "ID": ["A001", "A002", "A003", "A004", "A005",
           "A002", "A003", "A001", "A004", "A005",
           "A001", "A002", "A003", "A004",],
    
    "Project": ["Project 1", "Project 2", "Project 3", "Project 4", "Project 5",
                "Project 6", "Project 7", "Project 8", "Project 9", "Project 10",
                "Project 11", "Project 12", "Project 13", "Project 14"],
    
    "Cost": [100200, 200000, 450000, 550000, np.nan,
             600000, 400000, 350000, np.nan, 300000,
             200000, 100000, 300000, 200000],
    
    "Status": ["Finished", "Ongoing", "Finished", "Ongoing", "Finished",
               "Failed", "Finished", "Failed", "Ongoing", "Finished",
               "Failed", "Ongoing", "Finished", "Finished"]
}

# Create new DataFrame
df1 = pd.DataFrame(data)

# to make sure cost is numeric
df1["Cost"] = pd.to_numeric(df1["Cost"], errors='coerce')

# Handle first row if NaN
if pd.isna(df1.loc[0, "Cost"]):
    df1.loc[0, "Cost"] = df1.loc[1, "Cost"]

# Handle last row if NaN
if pd.isna(df1.loc[len(df1)-1, "Cost"]):
    df1.loc[len(df1)-1, "Cost"] = df1.loc[len(df1)-2, "Cost"]

# Loop through DataFrame to handle middle NaNs
for i in range(1, len(df1)-1):  # Avoiding first and last row
    if pd.isna(df1.loc[i, "Cost"]):
        prev_val = df1.loc[i-1, "Cost"]
        next_val = df1.loc[i+1, "Cost"]
        
        # Only fill if both neighbors are numbers
        if pd.notna(prev_val) and pd.notna(next_val):
            df1.loc[i, "Cost"] = (prev_val + next_val) / 2

print(df1)


```

```python
import pandas as pd

# Employee DataFrame
data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005'],
    'Name': ['John Alter', 'Alice Luxemburg', 'Tom Sabestine', 'Nina Adgra', 'Amy Johny'],
    'Gender': ['M', 'F', 'M', 'F', 'F'],
    'City': ['Paris', 'London', 'Berlin', 'Newyork', 'Madrid'],
    'Age': [25, 27, 29, 31, 30]
}

df2 = pd.DataFrame(data)

# adding first name and last name column using split
df2[['First Name', 'Last Name']] = df2['Name'].str.split(" ", n=1, expand=True)

# Removing the Name column by using drop
df2.drop(columns=['Name'], inplace=True)

print(df2.to_string(index=False))
```

```python
import pandas as pd

# Merging all 3 DataFrames on ID as it is the common value using merge based on ID 
Final = df1.merge(df2, on="ID").merge(df3, on="ID")

print(Final.to_string(index=False))
```

```python

Task 5
```

```python
# Creating new dataframe as Final and adding Bonus column
Final['Bonus'] = Final.apply(
    lambda row: row['Cost'] * 0.05 if row['Status'] == "Finished" else 0, #if statrus is Finished then giving 5% bonus on cost else 0%
    axis=1
)

print(Final.to_string(index=False))
```

```python

Task 6
```

```python
# Demote the designation level by 1, whose projects have status “fail”.
Final.loc[Final['Status'] == 'Failed', 'Designation Level'] -= 1

# Delete the employees record whose designation level is above 4.
Final = Final[Final['Designation Level'] <= 4]

print(Final.to_string(index=False))
```

```python

Task 7
```

```python
# adding “Mr.” and “Mrs.” to the first name column
Final['First Name'] = Final.apply(
    lambda row: f"Mr. {row['First Name']}" if row['Gender'] == 'M' else f"Mrs. {row['First Name']}",
    axis=1
)

# Drop Gender column
Final = Final.drop(columns=['Gender'])

print(Final.to_string(index=False))
```

```python

Task 8
```

```python
# Promote designation level by 1 for employees whose age > 29 using if condition
Final['Designation Level'] = Final.apply(
    lambda row: row['Designation Level'] + 1 if row['Age'] > 29 else row['Designation Level'],
    axis=1
)

print(Final.to_string(index=False))
```

```python

Task 9
```

```python
# Grouping by ID and First Name, then sum the Cost
TotalProjCost = Final.groupby(['ID', 'First Name'], as_index=False)['Cost'].sum()

# Renaming column
TotalProjCost.rename(columns={'Cost': 'Total cost'}, inplace=True)

print(TotalProjCost.to_string(index=False))
```
